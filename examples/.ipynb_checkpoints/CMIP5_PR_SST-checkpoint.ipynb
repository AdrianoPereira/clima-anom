```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import netCDF4
from netCDF4 import Dataset

from mpl_toolkits.axes_grid1 import make_axes_locatable

%run ../funtion/Funciones_v2.ipynb
```

```python
CMIP5_SST_path = '..\\Dados_CMIP5\\CMIP5_past1000_Tos\\tos_Omon_FGOALS-s2_past1000_r1i1p1_085001-184912.nc'
```

```python
time,lon,lat,SST_CMIP5 = read_netcdf(CMIP5_SST_path,'tos','lon','lat','time')
```

```python
SST_CMIP5[SST_CMIP5 == 1e+20] = 'NaN'
```

```python
print('Inicio: ',time[0])
print('Final : ',time[11995])
```

```python
SST = dictionary('SST',SST_CMIP5,len(lat),len(lon),1000)
```

```python
fig1 = plt.figure(figsize=(16,8),dpi=100)
m1 = Basemap(projection='cyl',llcrnrlat=-85,urcrnrlat=85,llcrnrlon=0,urcrnrlon=360,resolution='l')
m1.drawparallels(np.arange(-80,100,20),labels=[1,0,0,0],fontsize=12,color='gray',linewidth=0.5)
m1.drawmeridians(np.arange(-80,180,20),labels=[0,0,0,1],fontsize=12,color='gray',linewidth=0.5)
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

ax = plt.gca()

cmap = plt.cm.Spectral_r
a1 = plt.contourf(x,y,SST['jan']['data'][0,:,:],cmap=cmap,levels=None)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="4%", pad=0.05)

cbar1 = plt.colorbar(a1,cax=cax,ticks=None)

#plt.savefig(salida1,transparent = False,bbox_inches ='tight',pad_inches = 0.4)
```

```python
verano_anom = (SST['dec']['anom'][0:999,:,:] + SST['jan']['anom'][1:1000,:,:] + SST['feb']['anom'][1:1000,:,:])/3
np.shape(verano_anom)
```

```python
fig2 = plt.figure(figsize=(16,8),dpi=100)
m1 = Basemap(projection='cyl',llcrnrlat=-25,urcrnrlat=25,llcrnrlon=0,urcrnrlon=360,resolution='l')
m1.drawparallels(np.arange(-80,100,20),labels=[1,0,0,0],fontsize=12,color='gray',linewidth=0.5)
m1.drawmeridians(np.arange(-80,180,20),labels=[0,0,0,1],fontsize=12,color='gray',linewidth=0.5)
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

ax = plt.gca()

cmap = plt.cm.Spectral_r
a1 = plt.contourf(x,y,verano_anom[10,:,:],cmap=cmap,levels=None)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="4%", pad=0.05)

cbar1 = plt.colorbar(a1,cax=cax,ticks=None)

#plt.savefig(salida1,transparent = False,bbox_inches ='tight',pad_inches = 0.4)
```
