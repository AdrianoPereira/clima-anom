```python
import numpy as np
import clima_anom as ca
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
hgt_dir = '../data/Hgt_500hPa_Jan80_Dec19.nc'
```

```python
hgt_data = ca.read_netcdf(hgt_dir)
```

```python
lat = hgt_data['lat']
lon = hgt_data['lon']
```

```python
hgt_data['time'][0],hgt_data['time'][-1]
```

```python
hgt_dictionary = ca.clima_anom(hgt_data['hgt'][:,0,:,:])
```

```python
levels = np.arange(4600,6100,100)
cmap = plt.cm.Spectral_r

fig1 = plt.figure(figsize=(8,14),dpi=150)
m1 = Basemap(projection='cyl',llcrnrlat=min(lat),urcrnrlat=max(lat),llcrnrlon=min(lon),urcrnrlon=max(lon),resolution='l')

plt.subplot(6,2,1)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['jan']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('Janeiro Climatology')

plt.subplot(6,2,2)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['feb']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('February Climatology')

plt.subplot(6,2,3)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['mar']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('March Climatology')

plt.subplot(6,2,4)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['apr']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('April Climatology')

plt.subplot(6,2,5)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['may']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('May Climatology')

plt.subplot(6,2,6)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['jun']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('June Climatology')

plt.subplot(6,2,7)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['jul']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('July Climatology')

plt.subplot(6,2,8)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['ago']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('August Climatology')

plt.subplot(6,2,9)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['sep']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('September Climatology')

plt.subplot(6,2,10)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['oct']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('October Climatology')

plt.subplot(6,2,11)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['nov']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('November Climatology')

plt.subplot(6,2,12)
cmap = plt.cm.Spectral_r
m1.drawcoastlines(linewidth=0.5)
m1.drawcountries(linewidth=0.5)

lons,lats = np.meshgrid(lon,lat)
x,y = m1(lons,lats)

a1 = plt.contourf(x,y,hgt_dictionary['dec']['clim'][0,:,:],levels=levels,cmap=cmap)
plt.title('December Climatology')

cbaxes = fig1.add_axes([0.123, 0.1, 0.776, 0.01]) 
cb = plt.colorbar(a1,cax = cbaxes,orientation="horizontal")  

plt.savefig('../fig/Monthly_Climatology.png',transparent = False,bbox_inches ='tight',pad_inches = 0.4)
```
