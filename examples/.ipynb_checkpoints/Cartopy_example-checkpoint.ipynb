```python
import sys
import numpy as np
import pandas as pd
import clima_anom as ca
import cartopy
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
data_dir = '../data/wind_past_Jan8500_Dec1849.nc'
data = ca.read_netcdf(data_dir,2)
```

```python
time = data['time']
lon = data['lon']
lat = data['lat']
data = data['sfcWind'][0:96,:,:]

lon = lon +360
```

```python
lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
cmap=plt.cm.RdYlBu_r

plt.figure(figsize=(6,4),dpi=150)
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()

ax.set_extent([np.min(lon),np.max(lon),-22.3,22.3])

a1 = ax.contourf(lon2d, lat2d, data[0,:,:], cmap = cmap)

ax.gridlines(draw_labels=True, xlocs=np.arange(-180,180,30))
plt.colorbar(a1)
```
